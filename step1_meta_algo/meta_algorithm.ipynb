```python
# meta algo without learning
import numpy as np
from principal_meta import PrincipalMeta
from agent import Agent
from principal_agent_mdp import PrincipalAgentMDP
```

```python
GAMMA = 1

mdp = PrincipalAgentMDP(gamma=GAMMA)
agent = Agent(mdp)

# Principals reward function
r_p = np.zeros([mdp.n_states, mdp.n_outcomes])
for s in range(mdp.n_states):
    r_p[s, 0] = 14 / 9
    r_p[s, 1] = 0.0

# agent's reward already in MDP code


Q_bar_init = np.zeros((mdp.n_states, mdp.n_actions))

principal = PrincipalMeta(mdp, r_p, b_grid_step=0.1)
# Init
rho = {s: np.array([0.0, 0.0]) for s in range(mdp.n_states)} # for new agent

while True:

    # Inner Optimization: Agent's learning
    V_agent, Q_agent, pi_star = agent.solve(rho)
    print(f"Agent V: {V_agent}")

    # Outer Optimization:
    rho_old = {s: rho[s].copy() for s in range(mdp.n_states)}
    snapshots = principal.solve(agent)  # principal gets agent to have access to best responding agent.pi_star for each b 
    rho = principal.rho_star
    print(f"Contracts: {rho}")

    # algorithm converges if rho converges
    if all(np.allclose(rho[s], rho_old[s]) for s in range(mdp.n_states)):
        print("Meta-algorithm converged!")
        break

# Step 5: return (rho, pi)
print(f"\nFinal rho: {rho}")
print(f"Final pi:  {pi_star}")



    
```
